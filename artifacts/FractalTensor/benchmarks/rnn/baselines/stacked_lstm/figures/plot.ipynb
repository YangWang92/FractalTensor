```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set_theme(style='ticks', color_codes=True)
df = pd.read_table('for_plot.tsv', 
    header=0,
    sep='\t',
    usecols=['Test Name', 'Throughput'])

ax = sns.barplot(
    x='Test Name',
    y='Throughput',
    data=df,
    palette='ch:start=.2,rot=-.2')

ax.xaxis.label.set_visible(False)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
ax.yaxis.grid(True, clip_on=False)

ax.set_ylabel('Throughput(seq/sec)', fontsize=12)
ax.tick_params(labelsize=12)

plt.tight_layout()

fig = ax.get_figure()
fig.show()
fig.savefig('./stacked_lstm_impl.pdf', dpi=500, bbox_inches='tight')
```
