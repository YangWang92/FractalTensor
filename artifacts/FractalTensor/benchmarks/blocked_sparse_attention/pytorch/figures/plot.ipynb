```python
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='white', palette=None)

import matplotlib as mpl
mpl.rcParams['font.family'] = 'Times New Roman'
# mpl.rcParams['font.weight'] =  'bold'
# mpl.rcParams['savefig.dpi'] = 300
```

```python
df = pd.read_table(
    'bigbird_compute_vs_noncompute.tsv',
    header=0,
    sep='\t',
    usecols=[
        'test_name', 'compute', 'non_compute','total'
    ])
```

```python
f, ax = plt.subplots(figsize=(8, 5))

sns.set_color_codes('pastel')
sns.barplot(
    x='compute', y='test_name', data=df, label='Compute', color='lightskyblue')

sns.barplot(
    x='non_compute',
    y='test_name',
    data=df,
    label='Access',
    color='dodgerblue')

# ax.legend(ncol=1, loc='upper right', frameon=True, fontsize=16)
ax.legend(
    fontsize=20,
    frameon=True,
    bbox_to_anchor=(0.95, 1),
    shadow=True,
    facecolor='white')
ax.set(xlim=(0, 60), ylabel='', xlabel='')
sns.despine(left=True, bottom=True)

ax.set_ylabel(
    'sequence length, block size',
    fontdict={
        'fontsize': 20,
        'fontweight': 'bold'
    })
ax.set_xlabel('Duration(ms)', fontdict={'fontsize': 20, 'fontweight': 'bold'})
ax.tick_params(labelsize=18)

plt.tight_layout()
fig = ax.get_figure()
fig.savefig('bigbird_compute_vs_noncompute.pdf', dpi=600, bbox_inches='tight')
```
