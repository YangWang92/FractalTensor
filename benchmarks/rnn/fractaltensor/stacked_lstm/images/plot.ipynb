```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_table(
    'lstm_cell.csv',
    header=0,
    sep='|',
    usecols=['ExpName', 'Shape', 'Total(ms)'])

sns.set_theme(style='ticks', color_codes=True)
sns.set_style('whitegrid')
sns.set_context(
    'paper', rc={
        'font.size': 16,
        'axes.titlesize': 16,
        'axes.labelsize': 16
    })
```

```python
g = sns.catplot(
    data=df,
    kind='bar',
    x='Shape',
    y='Total(ms)',
    hue='ExpName',
    ci='sd',
    palette='mako',
    alpha=.6,
    height=6)
g.set_axis_labels('Shape([batch_size, hidden, length, depth])',
                  'Elapsed Time(ms)')
g.set_xticklabels(size=12)
g.set_yticklabels(size=12)

g._legend.remove()

plt.tight_layout()
plt.xticks(rotation=60)
plt.legend(
    bbox_to_anchor=(0.01, 0.99),
    loc='upper left',
    fontsize='16',
    title_fontsize='16')

fig = g.fig
fig.show()

fig.savefig('lstm_cell.png', dpi=500, bbox_inches='tight')
```
